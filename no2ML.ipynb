```python
df = pd.read_csv("midterm-regresi-dataset.csv")
```

```python
df_bersih = df.drop_duplicates()
```

```python
len(df_bersih)
```

```python
y = df_bersih.iloc[:, 0].astype(float)
X = df_bersih.iloc[:, 1:].astype(float)
```

```python
q1, q3 = np.percentile(y, [1, 99])
mask = (y >= q1) & (y <= q3)

X = X[mask]
y = y[mask]
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=70)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca = PCA(n_components=40)
X_pca = pca.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.2, random_state=42, shuffle=True
)
```

```python
import lightgbm as lgb

model = lgb.LGBMRegressor(
    device="gpu",
    boosting_type="gbdt",
    objective="rmse",
    n_estimators=8000,
    learning_rate=0.01,
    num_leaves=128,
    max_depth=-1,
    subsample=0.9,
    colsample_bytree=0.9,
    min_child_samples=5,
)
```

```python
model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    eval_metric="rmse"
)
```

```python
preds = model.predict(X_test)

mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))
r2 = r2_score(y_test, preds)

print("MAE :", mae)
print("RMSE:", rmse)
print("R²  :", r2)
```
